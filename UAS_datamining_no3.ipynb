```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
url = 'https://github.com/ekayuniar/datamining/raw/master/UAS/dataset%20soal%20no.3.xlsx'
data = pd.read_excel(url)
df = pd.DataFrame(data)

print(data.head())

print(data.columns)
```

```python
# Assume X and y are correctly defined based on your dataset structure
# For example:
X = data[['Usia', ' Kelahiran ke-', 'Waktu Kelahiran', 'Tekanan Darah',
       'Kelainan Jantung']]
y = data[' Cessar']

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define KNN model
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_scaled, y)

# Test data points
# Scenario 1: Ibu hamil dengan Usia 30 Tahun, Kelahiran ke-1, Waktu kelahiran sesuai HPL, Tekanan darah Normal
test_point_1 = np.array([[30, 1, 0, 1,0]])
# Scenario 2: Ibu hamil dengan Usia 29 Tahun, Kelahiran ke-2, Waktu kelahiran sesuai HPL, Tekanan darah Tinggi
test_point_2 = np.array([[29, 2, 0, 2,0]])

# Normalize test points
test_point_1_scaled = scaler.transform(test_point_1)
test_point_2_scaled = scaler.transform(test_point_2)

# Make predictions
prediction_1 = knn.predict(test_point_1_scaled)
prediction_2 = knn.predict(test_point_2_scaled)

print(f'Prediction for test point 1: {"Normal" if prediction_1 == 0 else "Caessarian"}')
print(f'Prediction for test point 2: {"Normal" if prediction_2 == 0 else "Caessarian"}')
```

```python
# Mendapatkan 5 tetangga terdekat untuk test_point_1 dan test_point_2
distances_1, indices_1 = knn.kneighbors(test_point_1_scaled, n_neighbors=5)
distances_2, indices_2 = knn.kneighbors(test_point_2_scaled, n_neighbors=5)

# Menampilkan tetangga terdekat dari data asli (belum dinormalisasi)
nearest_neighbors_1 = df.iloc[indices_1[0]]
nearest_neighbors_2 = df.iloc[indices_2[0]]

print("5 Tetangga terdekat untuk test point 1:")
print(nearest_neighbors_1)

print("\n 5 Tetangga terdekat untuk for test point 2:")
print(nearest_neighbors_2)
```

```python
print("Hasil prediksi untuk semua data:")
print("-" * 30)
for index, row in df.iterrows():
    prediction = knn.predict([row[X.columns]])[0]
    print("Data ke-", index+1)
    print("Usia:", row["Usia"])
    print("Kelahiran ke:", row[" Kelahiran ke-"])
    print("Waktu Kelahiran:", "HPL" if row["Waktu Kelahiran"] == 0 else ("Prematur" if row["Waktu Kelahiran"] == 1 else "Lewat HPL"))
    print("Tekanan Darah:", "Normal" if row["Tekanan Darah"] == 1 else ("Rendah" if row["Tekanan Darah"] == 0 else "Tinggi"))
    print("Kelainan Jantung:", "Normal" if row["Kelainan Jantung"] == 0 else "Kelainan")
    print("Persalinan:", "Normal" if row[" Cessar"] == 0 else "Caessarian")
    print("-" * 30)

print()
print("Hasil urutan data berdasarkan prediksi Caesarian knn:")
indexed_predictions = list(enumerate(df.apply(lambda row: knn.predict([row[X.columns]])[0], axis=1)))
sorted_predictions = sorted(indexed_predictions, key=lambda x: x[1])

for idx, pred in sorted_predictions:
    print(f"Data ke-{idx + 1}: Prediksi = {pred}")
```
